```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample accident data
data = {
    'City': ['Los Angeles', 'Houston', 'Phoenix', 'Chicago', 'Dallas', 'Austin', 'Atlanta', 'Miami', 'Seattle', 'Denver'],
    'Weather_Condition': ['Clear', 'Rain', 'Fog', 'Clear', 'Cloudy', 'Rain', 'Clear', 'Fog', 'Cloudy', 'Rain'],
    'Road_Condition': ['Dry', 'Wet', 'Wet', 'Dry', 'Dry', 'Wet', 'Dry', 'Wet', 'Dry', 'Wet'],
    'Time_of_Day': ['Night', 'Morning', 'Night', 'Evening', 'Morning', 'Night', 'Evening', 'Morning', 'Evening', 'Night'],
    'Accident_Count': [120, 80, 60, 90, 75, 50, 110, 55, 65, 70]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Print basic data
print("Sample Accident Data:")
print(df)

# 1. Bar plot: Accidents per city
plt.figure(figsize=(10, 5))
sns.barplot(x='City', y='Accident_Count', data=df, palette='magma')
plt.title('Accident Count by City')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Pie chart: Accidents by road condition
road_cond = df.groupby('Road_Condition')['Accident_Count'].sum()
plt.figure(figsize=(6, 6))
plt.pie(road_cond, labels=road_cond.index, autopct='%1.1f%%', colors=['orange', 'skyblue'])
plt.title('Accidents by Road Condition')
plt.axis('equal')
plt.show()

# 3. Bar plot: Accidents by weather
weather_cond = df.groupby('Weather_Condition')['Accident_Count'].sum().reset_index()
plt.figure(figsize=(8, 5))
sns.barplot(data=weather_cond, x='Weather_Condition', y='Accident_Count', palette='cool')
plt.title('Accidents by Weather Condition')
plt.show()

# 4. Line chart: Accidents by time of day
time_group = df.groupby('Time_of_Day')['Accident_Count'].sum().reset_index()
plt.figure(figsize=(7, 4))
sns.lineplot(data=time_group, x='Time_of_Day', y='Accident_Count', marker='o', color='green')
plt.title('Accidents by Time of Day')
plt.grid(True)
plt.show()
```
